```python
import pandas as pd

# 1. Lire le csv : survey_results_public.csv
df = pd.read_csv('./data/survey_results_public.csv')

# 2. Trouver le moyen d'afficher toutes les colonnes du DataFrame. Indice : pd.set_option(...)
pd.set_option('display.max_columns', None)

# 3. Afficher le type de chaque colonne :
print(df.dtypes)

# 4. Afficher le nombre de colonnes du DataFrame
print(df.shape[1])

# 5. Afficher le nombre d’employés à temps plein
fullTimeEmployeesNumber = df['Employment'].value_counts().get("Employed, full-time", 0)
print(fullTimeEmployeesNumber)

# 6. Afficher le pourcentage de personnes dont l'âge est compris entre 35 et 44 ans
numberOfPeopleAged35to44yo = df[df['Age'] == '35-44 years old']
percentage = (len(numberOfPeopleAged35to44yo) / len(df)) * 100
print(percentage)

# 7. Supprimer du DataFrame les colonnes suivantes :
# - Knowledge_1 à Knowledge_8
# - Frequency_1, Frequency_2, Frequency_3
for i in range(1, 9):
    colName = f"Knowledge_{i}"
    if colName in df.columns:
        df.drop(colName, axis=1, inplace=True)
        print(f"{colName} dropped")

for i in range(1, 4):
    colName = f"Frequency_{i}"
    if colName in df.columns:
        df.drop(colName, axis=1, inplace=True)
        print(f"{colName} dropped")


# 8. Afficher les enregistrements des personnes non employé qui ont un master
nonEmployedWithMastersDegree = df[(df['Employment'] == 'Not employed') & (df['EdLevel'] == 'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)')]
print(nonEmployedWithMastersDegree)

# 9. Calculer puis afficher la moyenne des salaires annuels de tous les répondants
averageSalaryAll = df['ConvertedCompYearly'].mean()
print(averageSalaryAll)

# 10. Calculer puis afficher la moyenne des salaires annuels des répondants français.
french = df[df['Country'] == 'France']
frenchAverageSalary = french['ConvertedCompYearly'].mean()
print(frenchAverageSalary)
```
